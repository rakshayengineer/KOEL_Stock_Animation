```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as pt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import random
import datetime
from itertools import count
from matplotlib.animation import FuncAnimation
%matplotlib notebook
df1 = pd.read_csv("KIRLOSENG.NS.csv")
df1["Date"] = pd.to_datetime(df1["Date"])
df1["Date"] = df1.Date.dt.date
df1["Close"] = df1.Close.apply(lambda x : round(x))
fig, ax = plt.subplots(figsize=(10,5))
plt.style.use(['dark_background'])
def animate(i):
    ax.clear()
    sns.lineplot(df1["Date"][:i+1], df1["Close"][:i+1],markers= True, dashes=False,ax=ax,legend=True)
    ax.set_title("Kirloskar Oil Engine Stock Journey from 1st April, 2022",fontdict={"fontsize" : 12},color = "white") 
    ax.set_xlabel("Date",fontdict={"fontsize" : 12},color = "orange")
    ax.set_ylabel("Stock Price",fontdict={"fontsize" : 12},color = "orange")
ani = FuncAnimation(fig, animate, frames = len(df1),interval = 1000) 
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import datetime
from matplotlib.animation import FuncAnimation,writers
%matplotlib notebook
df1 = pd.read_csv("KIRLOSENG.NS.csv")
df1["Date"] = pd.to_datetime(df1["Date"])
df1["Date"] = df1.Date.dt.date
df1["Close"] = df1.Close.apply(lambda x : round(x))
fig, ax = plt.subplots(figsize=(10,5))
plt.style.use(['dark_background'])
def animate(i):
    ax.clear()
    sns.lineplot(df1["Date"][:i+1], df1["Close"][:i+1],markers= True, dashes=False,ax=ax,legend=True)
    ax.set_title("Kirloskar Oil Engine Stock Journey from 1st April, 2022",fontdict={"fontsize" : 12},color = "white") 
    ax.set_xlabel("Date",fontdict={"fontsize" : 12},color = "orange")
    ax.set_ylabel("Stock Price",fontdict={"fontsize" : 12},color = "orange")
ani = FuncAnimation(fig, animate, frames = len(df1),interval = 1000)
Writer = writers["ffmpeg"]
writer = Writer(fps=10, metadata = {"artist" : "Me"}, bitrate = 1000)
ani.save("KOEL_Stock_Animation_1.mp4",writer)
```
